```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = "C:\\Users\\admin\\Desktop\\RF-Fingerprint-Extraction-for-Bluetooth\\BluetoothSignal\\signal\\channel39_1"
f = numpy.fromfile(open("file_path"), dtype=numpy.complex64)
```

```python
print(f.dtype)
```

```python
print("数据长度:", len(f))    # 确认长度是否合理（非零）
```

```python
print("前10个数据点:", f[:10])
```

## 时域波形（I/Q 分量）

```python
# 绘制前 1000 个点的实部（I）和虚部（Q）
plt.figure(figsize=(12, 6))
plt.plot(f.real[:1000], label="In-Phase (I)")
plt.plot(f.imag[:1000], label="Quadrature (Q)")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.title("I/Q Components (Time Domain)")
plt.legend()
plt.show()
```

## 幅度/相位波形

```python
# 计算幅度和相位
magnitude = np.abs(f)
phase = np.angle(f)

# 绘制幅度
plt.figure(figsize=(12, 4))
plt.plot(magnitude[:1000])
plt.xlabel("Sample Index")
plt.ylabel("Magnitude")
plt.title("Signal Magnitude")
plt.show()

# 绘制相位
plt.figure(figsize=(12, 4))
plt.plot(phase[:1000])
plt.xlabel("Sample Index")
plt.ylabel("Phase (radians)")
plt.title("Signal Phase")
plt.show()
```
